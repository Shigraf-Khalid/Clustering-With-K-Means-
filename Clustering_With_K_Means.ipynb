```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/cell_samples.csv")
df.head()
```

```python
plt.scatter(df.UnifShape,df['Clump'])
plt.xlabel('UnifShape')
plt.ylabel('Clump')
```

```python
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(df[['Clump','UnifShape']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]


plt.scatter(df1.Clump,df1['UnifShape'],color='green')
plt.scatter(df2.Clump,df2['UnifShape'],color='red')


plt.xlabel('UnifShape')
plt.ylabel('UnifShape')
plt.legend()
```

```python
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]


plt.scatter(df1.Clump,df1['Clump'],color='green')
plt.scatter(df2.Clump,df2['UnifShape'],color='red')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')

plt.xlabel('Clump')
plt.ylabel('UnifShape')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Clump']])
df['Clump'] = scaler.transform(df[['Clump']])

scaler.fit(df[['UnifShape']])
df['UnifShape'] = scaler.transform(df[['UnifShape']])
```

```python
df.head()
```

```python
plt.scatter(df.UnifSize,df['Clump'])
```

```python
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(df[['UnifShape','Clump']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]

plt.scatter(df1.UnifShape,df1['Clump'],color='green')
plt.scatter(df2.UnifShape,df2['Clump'],color='red')


plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```
